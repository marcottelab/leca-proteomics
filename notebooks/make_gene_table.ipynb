```python
import pandas as pd
import glob
import numpy as np
```

```python
og_file = '../ppi_ml/annotations/lists/leca_nogs.txt'
species_file = '../ppi_ml/data/meta/euk_codes_ordered.txt'
annot_file = '../ppi_ml/annotations/uniprot/leca_euNOGs_human-arath_annotated.tsv'
```

```python
leca_ogs = [line.strip() for line in open(og_file, 'r') if line.strip() != 'level_2759']
species = [line.strip() for line in open(species_file, 'r')]
an_df = pd.read_csv(annot_file, sep='\t')
```

```python
df_list = []
for code in species:
    file_dir = f'../ppi_ml/data/og_proteomes/nog_mapping'
    map_file = f'{file_dir}/{code}.euNOG.diamond.mapping.2759'
    df = pd.read_csv(map_file, sep='\t')
    df[code] = df['ProteinID'].str.extract('(?<=\|)(.*)(?=\|)', expand=False)
    for i in range(len(df)):
        if df[code][i] is np.nan:
            df[code][i] = df['ProteinID'][i] 
    df = df.groupby('ID', as_index=False).agg(lambda x: ', '.join(x.dropna()))
    df = df.drop('ProteinID', axis=1)
    df_list.append(df)
```

```python
from functools import reduce
gene_df = reduce(lambda x, y: pd.merge(x, y, on = 'ID', how = 'outer'), df_list)
gene_df
```

```python
out_df = pd.DataFrame()
out_df['ID'] = leca_ogs
out_df = out_df.merge(an_df, on='ID', how='left')
out_df = out_df.merge(gene_df, on='ID', how='left')
out_df
```
