```python
import pandas as pd
```

```python
pa_file = '../ppi_ml/results/coverage/og_coverage_by_species_phylo.150p.filtdollo.csv'
annot_file = '../ppi_ml/annotations/leca_eunog_annots_complete.030721.csv'
```

```python
pa_df = pd.read_csv(pa_file)
```

```python
all_species = pa_df.columns.values.tolist()
all_species.pop(0)
all_species
```

Conditions:
- all amorphea = 0
- tryb2 = 0
- tetts = 0
- euggr = 1 | 0
- phatc = 1 | 0
- plaf7 = 1 | 0
- plakh = 1 | 0
- plaba = 1 | 0
- all plants = 1 | 0

```python
not_in = pa_df.loc[:, 'dicdi':'human'].columns.to_list()
not_in.extend(['tryb2', 'tetts'])
not_in
```

```python
maybe_in = [s for s in all_species if s not in not_in]
maybe_in
```

```python
chloro_list = []
df = pa_df.copy()
for i in range(len(df)):
    not_in_vec = df[not_in].loc[i, :]
    not_sum = not_in_vec.sum()
    maybe_in_vec = df[maybe_in].loc[i, :]
    maybe_sum = maybe_in_vec.sum()
    if not_sum == 0 and maybe_sum > 0:
        chloro_list.append(df['ID'][i])
```

```python
len(chloro_list)
```

```python
chl_df = df[df['ID'].isin(chloro_list)]
chl_df = chl_df.reset_index(drop=True)
chl_df
```

```python
annot_df = pd.read_csv(annot_file)
out_df = chl_df.merge(annot_df, how='left', on='ID')
out_df.to_csv('../ppi_ml/results/coverage/candidate_chloro_ids_040323.csv', index=False)
```
