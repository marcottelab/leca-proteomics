```python
pdb_file = '../alphafold/lrrc40/results/lrrc40_lcp1_colab_unrelaxed_rank_1_model_3.pdb'
score_file = '../alphafold/lrrc40/results/lrrc40_lcp1_colab_unrelaxed_rank_1_model_3_scores.json'
```

```python
with open(score_file, 'r') as f:
    scores = json.load(f)
```
