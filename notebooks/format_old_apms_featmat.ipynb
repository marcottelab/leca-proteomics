```python
import pandas as pd
```

```python
fmat_file = '../ppi_ml/data/apms/featmat_humap'
fmat = pd.read_csv(fmat_file)
```

```python
cols = fmat.columns.tolist()
score_cols = cols[1:]
id_cols = ['ID1','ID2']
```

```python
fmat[['ID1','ID2']] = fmat['ID'].str.split(' ',expand=True)
fmat.drop('ID', axis=1, inplace=True)
```

```python
fmat = fmat[id_cols+score_cols]
```

```python
fmat.head()
```

```python
ogs_only = fmat[fmat['ID1'].str.startswith(('ENOG','KOG')) & fmat['ID2'].str.startswith(('ENOG','KOG'))]
```

```python
ogs_only.to_csv('../ppi_ml/data/apms/featmap_humap_euNOGs.csv', index=False, na_rep='')
ogs_only.to_pickle('../ppi_ml/data/apms/featmap_humap_euNOGs.pkl')
```
