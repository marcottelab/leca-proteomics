```python
import pandas as pd
import os
import re
```

```python
file_dir = '../ppi_ml/data/elutions/elut_ordered/'
file_pattern = '.*.unique.norm.ordered.elut'
```

```python
elut_files = [f for f in os.listdir(file_dir) if re.match(file_pattern, f)]
elut_files.remove('leca.unique.norm.ordered.elut')
elut_files
```

```python
leca_og_file = '../ppi_ml/annotations/leca_euNOGs_annotated.csv'
leca_df = pd.read_csv(leca_og_file)
```

```python
leca_nogs = set(leca_df['level_2759'].tolist())
```

```python
df = pd.read_csv(file_dir+elut_files[3])
```

```python
for f in elut_files:
    df = pd.read_csv(file_dir+f)
    df_filt = df[df['orthogroup'].isin(leca_nogs)]
    ext = f.split('.', 2)
    outfile = ext[0]+'.'+ext[1]+'.filtdollo.'+ext[2]
    df_filt.to_csv(file_dir+outfile, index=False)
```
