# map tetts IDs to leca NOGs

```python
def map_ids(x, d):
    try:
        new_id = d[x]
        return(new_id)
    except:
        return(x)
    
def fmt_xlinks(dfx):
    dfx = dfx[dfx['Crosslink Type'] == 'Inter']
    dfx['ID1'] = [map_ids(i, nog_dict) for i in dfx['Accession A']]
    dfx['ID2'] = [map_ids(i, nog_dict) for i in dfx['Accession B']]
    dfx.reset_index(inplace=True, drop=True)
    return(dfx)
    
def filter_leca(dfx):
    list_count = 0
    idx_list = []
    for i in range(len(dfx)):
        id1, id2 = dfx['ID1'][i], dfx['ID2'][i]
        try:
            if id1.startswith(tuple(nog_starts)) and id2.startswith(tuple(nog_starts)):
                idx_list.append(i)
        except AttributeError:
            list_count += 1
    print(f'# multi-matches = {list_count}')
    dfx_leca = dfx.iloc[idx_list]
    dfx_leca.reset_index(inplace=True, drop=True)
    return(dfx_leca)
```

```python
nog_starts = ['KOG', 'COG', 'ENOG']
```

```python
ppi_file = '../ppi_ml/results/ppi_predict/feature_sweep/100/scored_interactions_all_LinearSVC.csv'
cluster_file = '../ppi_ml/results/final/interactome_051023.csv'
```

```python
xlink_files = glob.glob('../xlink/data/raw/*csv')
xlink_files
```

```python
df_list = []
for f in xlink_files:
    exp = re.search('(?<=all42_).+(?=\.csv)', f)[0]
    print(exp)
    df = pd.read_csv(f)
    dfx = fmt_xlinks(df)
    dfx.to_csv(f'../xlink/data/processed/{exp}_nog_mapped.csv', index=False)
    dfx_leca = filter_leca(dfx)
    dfx_leca['type'] = exp
    df_list.append(dfx_leca)
```

```python
df_out = pd.concat(df_list)
df_out.to_csv('../xlink/results/leca_xlinks_062923.csv', index=False)
```

```python
df_out['fs'] = [frozenset({i,j}) for i,j in zip(df_out['ID1'], df_out['ID2'])]
```

```python
dfg = df_out.groupby(['Protein Descriptions A', 'Protein Descriptions B', 'fs'])
```

```python
counts = dfg.size().to_frame(name='counts')
exp_df = (counts
 .join(dfg.agg({'Max. XlinkX Score': 'mean'}).rename(columns={'Max. XlinkX Score': 'mean_score'}))
 .join(dfg.agg({'# CSMs': 'sum'}).rename(columns={'# CSMs': 'total_csms'}))
 .join(dfg['type'].apply(list))
 .sort_values('total_csms', ascending=False)
 .reset_index()
)

exp_df.to_csv('../xlink/results/leca_xlinks_summarized_062923.csv', index=False)
```

```python
clustering = pd.read_csv(cluster_file)
cut_cols = [col for col in clustering if col.startswith('cut')]
```

```python
clst_dict = dict()
for i in range(len(clustering)):
    prot = clustering['ID'][i]
    cluster = clustering['cut_199'][i]
    clst_dict[prot] = clustering[cut_cols].loc[i, :].values.flatten().tolist()
```

```python
def check_cluster(id1, id2, cdict):
    try:
        if cdict[id1][0] == cdict[id2][0]:
            intxn = set(cdict[id1]).intersection(set(cdict[id2]))
            return('TRUE', int(max(list(intxn))))
    except:
        return('FALSE', None)
    return('FALSE', None)
```

```python
ppis_xlinks[['same_cluster', 'cluster_num']] = [check_cluster(i, j, clst_dict) for i,j in ppis_xlinks['ID'].str.split(' ',expand=False)]
```
