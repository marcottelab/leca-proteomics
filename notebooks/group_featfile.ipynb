```python
import pandas as pd
import pickle
```

```python
fmat_file = '../ppi_ml/data/apms/humap/humap2_featmat_20200820.euNOGs.target_scores.pkl'
```

```python
with open(fmat_file, 'rb') as handle:
    fmat = pickle.load(handle)
```

```python
score_cols = fmat.columns[2:].tolist()
score_cols
```

```python
pair_count = fmat.groupby(['ID1', 'ID2']).size().sort_values(ascending=False)
```

```python
pair_count[0:10]
```

```python
kog1721 = fmat[(fmat['ID1'] == 'KOG1721') | (fmat['ID2'] == 'KOG1721')]
```

```python
kog1721.groupby(['ID1', 'ID2'])[score_cols].mean()
```

```python
fmat_grouped = fmat.groupby(['ID1', 'ID2'], as_index=False)[score_cols].mean()
```

```python
fmat_grouped
```

```python
fmat_grouped.to_csv('../ppi_ml/data/apms/humap/humap2_featmat_20200820.euNOGs.target_scores.group_mean.csv', index=False, na_rep='')
fmat_grouped.to_pickle('../ppi_ml/data/apms/humap/humap2_featmat_20200820.euNOGs.target_scores.group_mean.pkl')
```

```python
# in the format for the ML pipeline:
fmat_grouped['ID'] = fmat_grouped.ID1.str.cat(fmat_grouped.ID2, sep=' ')
fmat_grouped.drop(['ID1','ID2'], axis=1, inplace=True)
fmat_grouped = fmat_grouped[['ID'] + score_cols]
```

```python
fmat_grouped.to_csv('../ppi_ml/data/apms/featmap_humap', index=False, na_rep='')
fmat_grouped.to_pickle('../ppi_ml/data/apms/featmap_humap.pkl')
```
