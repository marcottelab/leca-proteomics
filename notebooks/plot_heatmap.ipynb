```python
# generate some random data
data = np.random.rand(25,25)

# create the heat map
plt.figure(figsize=(6, 6))

ax = sns.heatmap(
    data, vmin = 0, vmax = 1, 
    xticklabels=False, yticklabels=False,
    cmap="cividis_r", robust=True, cbar=False,
)
ax.set(xlabel="x axis", ylabel="y axis")
ax.xaxis.tick_top()
ax.xaxis.set_label_position("top")

plt.arrow(x=0, y=0, dx=0, dy=25, length_includes_head=True, shape="full", color="black", 
          width=0.25, head_width=1, head_length=1.25, clip_on=False)
plt.arrow(x=0, y=0, dx=25, dy=0, length_includes_head=True, shape="full", color="black", 
          width=0.25, head_width=1, head_length=1.25, clip_on=False)

# show the plot
plt.show()
```

```python
thres = 150
ifile = f'leca.unique.filtdollo.{thres}p.norm.ordered.clst.correlation.average.elut.pkl'
path = '../ppi_ml/results/elut_clustering/'
df = pd.read_pickle(path+ifile)
df.set_index('orthogroup', inplace=True)
df.head()
```

```python
x = df.to_numpy()
num_ogs = len(x)
x.shape[1]
```

```python
plt.figure(figsize=(12, 6))

ax = sns.heatmap(
    x, xticklabels=False, yticklabels=False,
    cmap="cividis_r", robust=True, cbar=False,
)
ax.set(xlabel=f'10,481 MS fractions across 31 eukaryotes', ylabel=f'{num_ogs:,} LECA orthogroups')
ax.xaxis.tick_top()
ax.xaxis.set_label_position("top")

w=40
plt.arrow(x=-1, y=0, dx=0, dy=x.shape[0], length_includes_head=True, shape="full", color="black", 
          clip_on=False, width=w, head_width=w*4, head_length=w*5)
plt.arrow(x=0, y=-1, dx=x.shape[1], dy=0, length_includes_head=True, shape="full", color="black", 
          clip_on=False, width=w, head_width=w*4, head_length=w*5)

plt.savefig(f'../ppi_ml/figures/elut_heatmap_{thres}p_filtdollo_norm_rocket.pdf', 
           dpi=300, transparent=True)
plt.savefig(f'../ppi_ml/figures/elut_heatmap_{thres}p_filtdollo_norm_rocket.png', 
           dpi=300, transparent=True)

plt.show()
```
