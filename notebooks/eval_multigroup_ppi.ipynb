```python
import pandas as pd
import pickle
```

```python
def merge(lsts):
    sets = [set(lst) for lst in lsts if lst]
    merged = True
    while merged:
        merged = False
        results = []
        while sets:
            common, rest = sets[0], sets[1:]
            sets = []
            for x in rest:
                if x.isdisjoint(common):
                    sets.append(x)
                else:
                    merged = True
                    common |= x
            results.append(common)
        sets = results
    return sets
```

```python
labeled_file = '../ppi_ml/data/featmats/featmat_final_labeled_traintest.pkl'
```

```python
with open(labeled_file, 'rb') as handle:
    ldf = pickle.load(handle)
```

```python
ldf.head()
```

```python
index_list = []
multi_count = 0
for i in range(len(ldf)):
    group = ldf['group'][i]
    if type(group) == list:
        index_list.append(i)
        multi_count += 1
```

```python
multi = ldf.iloc[index_list]
len(multi)
```

```python
multi.head()
```

```python
mgroup_list = multi['group'].tolist()
```

```python
merged = merge(mgroup_list)
```

```python
len(merged)
```

```python
super_grp_num = 1
super_grp_dict = dict()
for group in merged:
    for old_gnum in group:
        super_grp_dict[old_gnum] = super_grp_num
    super_grp_num += 1
```
